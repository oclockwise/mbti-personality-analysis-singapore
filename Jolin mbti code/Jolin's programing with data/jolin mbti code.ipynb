##### **1.3.4 Limitations and constraints of the data**

MBTI and its results may not always be dependable.

- **Sampling Bias:** The reliability of MBTI results can be affected if certain groups, like people with multiple accounts or those really interested in personality tests, are overrepresented in the data. This can make the results not truly reflect the whole population.

- **Incomplete Data:** If we don't have access to enough different sources of data about MBTI, the results might not tell the whole story. For instance, if there's not much data available in Singapore about MBTI as it is not a popular topic here, our analysis might be missing important information, which could make it less reliable. Specific information that might be missing includes age, gender, location, and career.

- **Test-Retest Reliability:** Sometimes, when people take the MBTI test again and again, they can get different results each time. This makes it hard to trust the test's consistency over time.

- **Self-Selection Bias:** People who take the MBTI test do so by choice. This means the results might be skewed towards those already interested in personality types.

- **Self-Reporting Issues:** The MBTI relies on how people report their traits. Factors like mood, honesty, and understanding of the questions can affect the accuracy of these reports.

- **Cultural Bias:** The MBTI was designed within a specific culture. It might not fully capture personality traits across different cultures, which can limit its applicability to Singaporeans.

- **Limited Representation:** Not everyone can easily access or complete the MBTI test. This includes individuals without internet access or those who don't speak the language well, which may lead to an unrepresentative sample.

#### **2.1 Web Page Accessibility and Data Extraction**

Before proceeding with the web scraping project, it's crucial to ensure that the target webpages are accessible. In this section, we describe the function developed to verify the accessibility of webpages and retrieve their content if accessible. 
We identify the webpage URL from which we want to retrieve data. Having the correct URL to access the desired resource effectively is crucial. In our case, we are interested in retrieving data from the country profile page of Singapore on the 16Personalities website at "https://www.16personalities.com/country-profiles/singapore"

```python
def top10webpage(url):
    # Send a GET request
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse HTML content
        soup = BeautifulSoup(response.content, "html.parser")
        return soup
    else:
        print("Failed to retrieve data. Status code:", response.status_code)
        return None

url = "https://www.16personalities.com/country-profiles/singapore"
soup = top10webpage(url)
```

##### **2.2.1 Standardizing data format**

###### **2.2.1.1 Using JSON for Extraction, Cleaning, and Parsing of JSON-like Strings**


The data we require is embedded within the HTML structure of the webpage, specifically within the :collection attribute of the <country-profiles-top-ten-list> element. This JSON-like string contains valuable information about the personality types that we want to analyze further. However, the data retrieved from the webpage is in a JSON-like format, and it contains HTML encoding such as &quot; instead of standard double quotes ("). Without extracting this string, we cannot access the raw data needed for analysis. To properly interpret this data as JSON and work with it in Python, we need to ensure that the string is formatted correctly. Therefore, we extract the JSON-like string and convert it to a proper JSON string. Once the JSON-like string is cleaned and formatted correctly, we parse it into a Python dictionary to manipulate and analyze the data.

```python
# Define the component variable
component = soup.find("country-profiles-top-ten-list")

# Extract the JSON-like string
json_data = component.attrs[":collection"]

# Convert the JSON-like string to a proper JSON string
json_data = json_data.replace('&quot;', '"')

# Parse the JSON string
data = json.loads(json_data)
print(data)
```

Now, the raw data from the webpage has been converted into a structured format that we can effectively work with in Python. 

##### **2.2.2 Remove duplicates and aggregating dataset**

After retrieving the data, it is observed that some personalities are repeated twice within the dataset. This duplication arises from the separation of personalities into Assertive and Turbulent categories, as seen with examples like Mediator-A and Mediator-T. We intend to merge them with their corresponding percentages for this analysis.
We start by creating a dictionary to hold the aggregated percentages for each personality type. This dictionary will iterate through each item in the data list, which contains the personality data. For each item, we extract the name of the personality and its associated percentage. The name is then cleaned by removing the substrings "Assertive " and "Turbulent". After cleaning, the name and percentage are added to the dictionary. If the personality already exists in the dictionary, its percentage is incremented by the new percentage value. This process ensures that each personality type is represented once in the aggregated data, eliminating any duplicates caused by the presence of "Assertive" and "Turbulent" labels.
The aggregated data are then sorted in descending order based on their percentage.


```python
# Aggregate the percentages by name
aggregated_data = defaultdict(float)
for item in data:
    name = item['name'].replace("Assertive ", "").replace("Turbulent ", "")
    percentage = float(item['percentage'])
    aggregated_data[name] += percentage

# Sort the aggregated data by percentage in descending order
sorted_data = sorted(aggregated_data.items(), key=lambda x: x[1], reverse=True)

# Print the aggregated and sorted data with ranking
for rank, (name, percentage) in enumerate(sorted_data, start=1):
    print(f"Rank {rank}: Name: {name}, Percentage: {percentage:.2f}%")
```

We utilize a dictionary named 'name_to_code' to map each personality name to its corresponding code. This dictionary contains key-value pairs where the personality names ('Mediator', 'Campaigner', etc.) are mapped to their respective codes ('INFP', 'ENFP', etc.)


```python
# Dictionary to map names to codes
name_to_code = {
    'Mediator': 'INFP',
    'Campaigner': 'ENFP',
    'Defender': 'ISFJ',
    'Adventurer': 'ISFP',
    'Consul': 'ESFJ',
    'Entertainer': 'ESFP',
    'Advocate': 'INFJ',
    'Protagonist': 'ENFJ',
    'Logician': 'INTP',
    'Logistician': 'ISTJ',
    'Executive': 'ESTJ',
    'Debater': 'ENTP',
    'Virtuoso': 'ISTP',
    'Entrepreneur': 'ESTP',
    'Architect': 'INTJ',
    'Commander': 'ENTJ'
}
# Create a DataFrame from the sorted data
df = pd.DataFrame(sorted_data, columns=['Name', 'Percentage'])

# Update the Name column using the dictionary
df['Code'] = df['Name'].map(name_to_code)
df
```

We can gather the combined percentages for each of the personality abbreviations (I, E, N, S, F, T, J, P) by looping through the data and aggregating the percentages accordingly. 

```python
# Aggregate the percentages by the four traits
aggregated_data = defaultdict(float)

for index, row in df.iterrows():
    code = row['Code']
    percentage = row['Percentage']
    
    for char in code:
        aggregated_data[char] += percentage

# Create a DataFrame from the aggregated data
agg_df = pd.DataFrame(list(aggregated_data.items()), columns=['trait', 'Total Percentage'])
# Save the DataFrame to a CSV file for comparison analysis later on
agg_df.to_csv('agg_df.csv', index=False)

agg_df
```

Now that we have the combined percentages for each personality trait (I, E, N, S, F, T, J, P), we can further categorize these traits into their respective pairs: I/E (Introversion/Extraversion), S/N (Sensing/Intuition), F/T (Feeling/Thinking), and J/P (Judging/Perceiving) for comparison.

```python
from collections import defaultdict

# Initialize the defaultdicts for each trait
aggregated_data_IE = defaultdict(float)
aggregated_data_NS = defaultdict(float)
aggregated_data_FT = defaultdict(float)
aggregated_data_JP = defaultdict(float)

# Aggregate the percentages for each trait
for index, row in df.iterrows():
    code = row['Code']
    percentage = row['Percentage']
    
    # Aggregate for 'I/E' trait
    if code[0] in ['I', 'E']:
        aggregated_data_IE[code[0]] += percentage
    
    # Aggregate for 'N/S' trait
    if code[1] in ['N', 'S']:
        aggregated_data_NS[code[1]] += percentage
    
    # Aggregate for 'F/T' trait
    if code[2] in ['F', 'T']:
        aggregated_data_FT[code[2]] += percentage
    
    # Aggregate for 'J/P' trait
    if code[3] in ['J', 'P']:
        aggregated_data_JP[code[3]] += percentage

```

Additionally, we can map these traits to their role groups and aggregate their percentages accordingly.

```python
# Define the mappings of traits to role groups
role_mapping = {
    'NT': ['N', 'T'],
    'NF': ['N', 'F'],
    'SJ': ['S', 'J'],
    'SP': ['S', 'P']
}

# Initialize defaultdicts to aggregate percentages for each role group
aggregated_roles = defaultdict(float)

# Aggregate the percentages for each role group
for role, traits in role_mapping.items():
    for trait in traits:
        aggregated_roles[role] += aggregated_data[trait]

# Calculate the total percentage for all role groups
total_percentage_roles = sum(aggregated_roles.values())
```

##### **2.2.3 Normalizing**

Even though we might have retrieved the respective information for each pair, we need to normalize it. Normalization ensures that the percentages for each trait are relative and comparable. Without normalization, the raw percentages might be misleading because they are not scaled appropriately.

```python
# Normalize the total percentages for each trait
total_percentage_IE = sum(aggregated_data_IE.values())
normalized_data_IE = {key: round((value / total_percentage_IE) * 100, 1) for key, value in aggregated_data_IE.items()}
agg_df_IE = pd.DataFrame(list(normalized_data_IE.items()), columns=['trait', 'Normalized Percentage'])

total_percentage_NS = sum(aggregated_data_NS.values())
normalized_data_NS = {key: round((value / total_percentage_NS) * 100, 1) for key, value in aggregated_data_NS.items()}
agg_df_NS = pd.DataFrame(list(normalized_data_NS.items()), columns=['trait', 'Normalized Percentage'])

total_percentage_FT = sum(aggregated_data_FT.values())
normalized_data_FT = {key: round((value / total_percentage_FT) * 100, 1) for key, value in aggregated_data_FT.items()}
agg_df_FT = pd.DataFrame(list(normalized_data_FT.items()), columns=['trait', 'Normalized Percentage'])

total_percentage_JP = sum(aggregated_data_JP.values())
normalized_data_JP = {key: round((value / total_percentage_JP) * 100, 1) for key, value in aggregated_data_JP.items()}
agg_df_JP = pd.DataFrame(list(normalized_data_JP.items()), columns=['trait', 'Normalized Percentage'])
```

```python
# Normalize the percentages for each role group
agg_df_roles = pd.DataFrame(list(aggregated_roles.items()), columns=['Role Group', 'Total Percentage'])
agg_df_roles['Normalized Percentage'] = round((agg_df_roles['Total Percentage'] / total_percentage_roles) * 100, 1)
```

##### **3.2.1 Removing Irrelevant Information**

We utilize the re.sub() function from Python's re module to clean the comments. This function allows us to remove unwanted elements such as URLs, subreddit mentions (like "r/subreddit"), and non-alphanumeric characters from the text data. This preprocessing step ensures that our text data is ready for sentiment analysis, where the presence of irrelevant information could skew the results.

```python
# Function to remove irrelevant information from a comment
def remove_irrelevant_info(comment):
    return re.sub(r'http\S+|r/[^\s]+|u/[^\s]+|[^a-zA-Z0-9\s]', '', comment)

# http\S+: Removes URLs starting with "http".
# r/[^\s]+: Removes subreddit mentions in the format "r/subreddit".
# u/[^\s]+: Removes user mentions in the format "u/username".
# [^a-zA-Z0-9\s]: Removes non-alphanumeric characters, leaving only letters, digits, and spaces.
```

##### **3.2.2 Tokenizing**

We employs NLTK's word_tokenize() method to tokenize or break down a comment into individual words or tokens. This is a crucial step in natural language processing (NLP) tasks as it allows us to analyze text data at a more granular level. This process enables us to treat each word as a separate entity, facilitating further analysis.

```python
# Function to tokenize a comment using NLTK
def tokenize_comment(comment):
    tokenized_comment = word_tokenize(comment)
    return tokenized_comment
```

##### **3.2.3 Stemming**

We use stemming to standardize words in our subreddit comments. This simplifies the text data by reducing words to their root forms, such as "running" and "runner" becoming "run". This standardization facilitates analysis and comparison of the text data.

```python
# Function to apply stemming using NLTK's PorterStemmer
def stem_words(words):
    ps = PorterStemmer()
    return [ps.stem(word) for word in words]
```

#### **3.3 Generating features**

##### **3.3.1 Sentiment Analysis**

We used NLTK's Sentiment Intensity Analyzer to evaluate the sentiment of the cleaned and processed comments. Each comment was assigned a sentiment score, which was then recorded along with its MBTI type and the original comment. For this analysis, we will be searching for 10 comments in each MBTI type subreddit. This is to ensure that we have a diverse and representative sample of comments for each personality type. By analyzing these samples, we can draw more accurate conclusions about the overall sentiment and common traits of each MBTI type.

```python
# Function to search for comments in each MBTI type subreddit
def search_comments(subreddit_name):
    try:
        subreddit = reddit.subreddit(subreddit_name)
        comment_count = 0
        for submission in subreddit.new(limit=10):
            submission.comments.replace_more(limit=None)
            for comment in submission.comments.list():
                if comment_count >= 10:
                    break
                comment_count += 1
               
    except Exception as e:
        print(f"Error fetching comments for subreddit r/{subreddit_name}: {e}")
```

```python
# Function to perform sentiment analysis using NLTK's SentimentIntensityAnalyzer
def analyze_sentiment(comment):
    scores = sia.polarity_scores(comment)
    return scores['compound']

# Initialize DataFrame to store sentiment analysis results
sentiment_df = pd.DataFrame(columns=['MBTI Type', 'Comment', 'Sentiment Score'])

# Function to search for comments in each MBTI type subreddit
def search_comments(subreddit_name):
    try:
        subreddit = reddit.subreddit(subreddit_name)
        comment_count = 0
        for submission in subreddit.new(limit=10):
            submission.comments.replace_more(limit=None)
            for comment in submission.comments.list():
                if comment_count >= 10:
                    break
                comment_body = comment.body
                cleaned_comment = remove_irrelevant_info(comment_body)
                tokenized_comment = tokenize_comment(cleaned_comment)
                stemmed_comment = stem_words(tokenized_comment)
                sentiment_score = analyze_sentiment(' '.join(stemmed_comment))
                sentiment_df.loc[len(sentiment_df)] = [subreddit_name, comment_body, sentiment_score]
                comment_count += 1
    except Exception as e:
        print(f"Error fetching comments for subreddit r/{subreddit_name}: {e}")

# Search for comments and perform sentiment analysis for each MBTI type subreddit
for mbti_type in mbti_types:
    search_comments(mbti_type)
```

```python
# Group sentiment DataFrame by MBTI type and calculate mean sentiment score
average_sentiment_per_mbti = sentiment_df.groupby('MBTI Type')['Sentiment Score'].mean()

# Print the average sentiment score for each MBTI type
print("Average Sentiment Score for Each MBTI Type:")
print(average_sentiment_per_mbti)
```

This will help us further analyze and understand the common traits shared by individuals within each MBTI type. By examining these frequent words, we can gain deeper insights into the unique characteristics and perspectives of each personality group.

```python
# Define the list of stopwords
stop_words = set(stopwords.words('english'))

# Filter the DataFrame for ESFP comments
esfp_comments = sentiment_df[sentiment_df['MBTI Type'] == 'ESFP']['Comment']

# Initialize a list to store all words for ESFP
all_esfp_words = []

# Process each ESFP comment to extract words
for comment in esfp_comments:
    cleaned_comment = remove_irrelevant_info(comment)
    tokenized_comment = tokenize_comment(cleaned_comment)
    stemmed_comment = stem_words(tokenized_comment)
    all_esfp_words.extend(stemmed_comment)

# Remove stopwords for ESFP
filtered_esfp_words = [word for word in all_esfp_words if word not in stop_words]

```
